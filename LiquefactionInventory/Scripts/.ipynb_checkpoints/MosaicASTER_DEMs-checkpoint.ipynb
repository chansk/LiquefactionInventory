```python
# Import system modules
import pandas as pd 
import os
import arcpy
```

```python
# Change working directory
os.chdir("C:\\Users\\achans01\\Downloads\\WatBods\\WatBodsCombined")
```

```python
path_name = "C:\\Users\\achans01\\Downloads\\WatBods\\WatBodsCombined"
aster_list = os.listdir(path_name)
```

```python
aster_list
```

```python
num_aster = len(aster_list)
num_aster
```

```python
aster_tifs = []
for i in range(0,num_aster):
    aster_tifs.append(path_name + "\\" + aster_list[i]+"\\"+aster_list[i]+"_dem.tif")
print(aster_tifs)
```

```python
output_name = "Aster_Mosaic_DEMs.tif"
```

```python
arcpy.MosaicToNewRaster_management(aster_tifs, "C:\\Users\\achans01\\Downloads\\WatBods", output_name, "", "16_BIT_SIGNED", "", "1", "MEAN", "FIRST")
```
